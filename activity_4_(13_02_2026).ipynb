```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Upload image
uploaded = files.upload()

# Get filename automatically
image_path = list(uploaded.keys())[0]

# Read image
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if img is None:
    print("Image not loaded. Try again.")
else:
    print("Image loaded successfully!")
```

```python
zoom_levels = ["High", "Medium-High", "Medium", "Low", "Very Low"]
scales = [1.0, 0.75, 0.5, 0.25, 0.125]

means = []
variances = []

for scale in scales:
    resized = cv2.resize(img, None, fx=scale, fy=scale, interpolation=cv2.INTER_AREA)
    flat = resized.flatten()

    means.append(np.mean(flat))
    variances.append(np.var(flat))

print("Means:", means)
print("Variances:", variances)

plt.figure(figsize=(8,5))
plt.plot(zoom_levels, variances, marker='o')
plt.xlabel("Zoom Level")
plt.ylabel("Variance")
plt.title("Zoom Level vs Variance")
plt.xticks(rotation=45)
plt.show()
```
