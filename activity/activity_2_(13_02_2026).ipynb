```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def create_blob(radius, size=400):
    img = np.zeros((size, size), dtype=np.uint8)
    cv2.circle(img, (size//2, size//2), radius, 255, -1)
    return img
```

```python

def apply_log(image, sigma):
    ksize = int(6 * sigma + 1)
    if ksize % 2 == 0:
        ksize += 1

    blurred = cv2.GaussianBlur(image, (ksize, ksize), sigma)
    log = cv2.Laplacian(blurred, cv2.CV_64F)
    log = cv2.convertScaleAbs(log)
    return log

# Blob sizes (small → large)
blob_sizes = [20, 40, 80, 120]
sigmas = [1, 2, 4, 8]
```

```python
for i in range(len(blob_sizes)):

    blob = create_blob(blob_sizes[i])
    log_output = apply_log(blob, sigmas[i])

    print(f"\nBlob Radius = {blob_sizes[i]}")
    cv2_imshow(blob)

    print(f"LoG Filter Sigma = {sigmas[i]}")
    cv2_imshow(log_output)

print("Activity Complete ✅")
```
