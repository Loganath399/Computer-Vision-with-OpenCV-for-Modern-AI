```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()

# Get filename
image_path = list(uploaded.keys())[0]
```

```python
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

level0 = gray
level1 = cv2.pyrDown(level0)   # Half size
level2 = cv2.pyrDown(level1)   # Quarter size
level3 = cv2.pyrDown(level2)   # One-eighth size

pyramid = [level0, level1, level2, level3]
```

```python
level0 = gray
level1 = cv2.pyrDown(level0)   # Half size
level2 = cv2.pyrDown(level1)   # Quarter size
level3 = cv2.pyrDown(level2)   # One-eighth size

pyramid = [level0, level1, level2, level3]
```

```python
edges = []
for level in pyramid:
    edge = cv2.Canny(level, 100, 200)
    edges.append(edge)
```

```python
titles = [
    "Level 0 - Original",
    "Level 1 - Half Size",
    "Level 2 - Quarter Size",
    "Level 3 - One-Eighth Size"
]

plt.figure(figsize=(12, 10))

for i in range(4):
    # Show pyramid image
    plt.subplot(4, 2, 2*i+1)
    plt.imshow(pyramid[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

    # Show edges
    plt.subplot(4, 2, 2*i+2)
    plt.imshow(edges[i], cmap='gray')
    plt.title("Canny Edges")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
